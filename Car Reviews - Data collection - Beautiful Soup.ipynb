## Beautiful Soup - Autocarindia

```python
import sys
import bs4 as bs
import csv
import requests
import time
try:
    from urllib.request import urlopen
except:
    from urllib2 import urlopen
import pandas as pd
import time
import os
import csv
from random import randint
from time import sleep
```

## Extract reviews from AutocarIndia

```python
cars = []
final_views = []
years= []

for i in range(45):
    if (i == 0):
        url_link = "https://www.autocarindia.com/car-reviews/"
    else:
        url_link = "https://www.autocarindia.com/car-reviews/%d" % i
    print (url_link)

    sleep(randint(1,2))
    page = urlopen(url_link)
    soup_obj = bs.BeautifulSoup(page, "lxml")
    
    # Find the time of the post to extract model year approx since model year is missing lot of times
    for t in soup_obj.find_all('time'):
        y = t.text.strip().split(" ")[2]
        if "ago" not in y:
            years.append(y)
        else:
            years.append("")
        
    # Find the title of the review and extract Manufacturer, Model
    alltxt = soup_obj.find_all('a')
    for t in alltxt:
        try:
            title = t.get("title")
            if "test drive" in title:
                cars.append(title.split(",")[0])                 
        except:
            #print ("error")
            pass

    # Find the number of views of the review
    views = soup_obj.find_all('ul', {"class" : "action"})
    for view in views:
        litags = view.find_all("li")
        if len(litags) >= 1:
            final_views.append(litags[0].text)
        else:
            pass                    
```

## Extract Model and Manufacturer details

```python
manufacturer = []
models = []
fuel_types = []
car_info = []
avl_fueltypes = ["petrol", "diesel", "hybrid", "electric"]
for idx, c in enumerate(cars):
    print (c)
    values = c.split(" ") 
    fuel_type = ""
    for v in values:
        for f in avl_fueltypes:
            if f in v.lower():
                #print (v)
                fuel_types.append(v)
                fuel_type = v
        
    if values[0].isdecimal() == True: # check if decimal
        print ("%%%", c)
        if "New" in values[1] or "Upcoming" in values[1]:
            manufacturer.append(values[2])
            manuf = values[2]
            models.append(values[3])
            model = values[3]
        else:
            manufacturer.append(values[1])
            manuf = values[1]
            models.append(values[2])
            model = values[2]   
    else:
        print ("===", c)
        if "New" in values[0] or "Upcoming" in values[0]:
            if values[1].isdecimal():
                print (c)
                manufacturer.append(values[2])
                manuf = values[2]
                models.append(values[3])
                model = values[3]
            else:
                print (c)
                manufacturer.append(values[1])
                manuf = values[1]
                models.append(values[2])
                model = values[2]
        elif "Maruti" in values[0]:
            if "Suzuki" in values[1]:
                manufacturer.append(values[2])
                manuf = values[2]
                models.append(values[3])
                model = values[3]   
            else:
                manufacturer.append(values[1])
                manuf = values[1]
                models.append(values[2])
                model = values[2]                   
        else:
            manufacturer.append(values[0])
            manuf = values[0]
            models.append(values[1])
            model = values[1]   

    if "Merc" in manuf:
        manuf = "Mercedes"
    print ("=========", manuf)
    car_info.append((manuf.strip(), model.strip(), fuel_type.strip(), years[idx].strip(), final_views[idx].strip().split(" ")[0].strip()))
```

```python
car_review_df = pd.DataFrame(car_info, columns = ["Manufacturer","Model","Fuel Type","Model Year","Number of Views"])
```

```python
car_review_df.to_csv("Car Reviews.csv", index=False)
```
